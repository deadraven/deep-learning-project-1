```python
import os
import numpy as np
import matplotlib.pyplot as plt

from sklearn.utils import shuffle

from tensorflow import keras
```

```python
import cv2
```

```python
Dir_1="/content/drive/MyDrive/untitled folder/Covid19-dataset/test"
Dir_2="/content/drive/MyDrive/untitled folder/Covid19-dataset/train"
CAT=["Covid","Normal","Viral Pneumonia"]
```

```python
test=[]
for cat in CAT:
    path=os.path.join(Dir_1,cat)
    class_num=CAT.index(cat)
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        new_array=cv2.resize(img_array,(500,500))
        test.append([new_array,class_num])
```

```python
print(len(test))
```

```python
train=[]
for cat in CAT:
    path=os.path.join(Dir_2,cat)
    class_num=CAT.index(cat)
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        new_array=cv2.resize(img_array,(500,500))
        train.append([new_array,class_num])
```

```python
print(len(train))
```

```python
import random
```

```python
random.shuffle(test)
random.shuffle(train)
```

```python
x_train=[]
y_train=[]
x_test=[]
y_test=[]
```

```python
for features,label in train:
    x_train.append(features)
    y_train.append(label)
for features,label in test:
    x_test.append(features)
    y_test.append(label)
```

```python
x_train=(np.array(x_train)).reshape(-1,500,500,1)
x_test=(np.array(x_test)).reshape(-1,500,500,1)
```

```python
x_train[0]
```

```python
x_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
```

```python
x_train=x_train/255

x_test=x_test/255
```

```python
model = Sequential()

model.add(Conv2D(256, (3, 3), input_shape=[500,500,1]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(256, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dense(3))
model.add(Activation('softmax'))
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
y_train=np.array(y_train).reshape(-1,1)
y_test=np.array(y_test).reshape(-1,1)
```

```python
encoder = OneHotEncoder(sparse_output=False)
y_train = encoder.fit_transform(y_train.reshape(-1,1))
y_test = encoder.fit_transform(y_test.reshape(-1,1))
```

```python
y_test.shape
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, batch_size=4, epochs=15)
```

```python
model.evaluate(x_test,y_test,batch_size=1)
```
